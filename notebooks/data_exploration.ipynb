## Step 3: Explore All Sheets in Grossi Excel

```python
# Check if Excel file has multiple sheets
grossi_excel = data_dir / 'Grossi_et_al._Piloted_data_extraction_strategy.xlsx'

if grossi_excel.exists():
    xls = pd.ExcelFile(grossi_excel)
    print(f"Number of sheets: {len(xls.sheet_names)}")
    print(f"\nSheet names:")
    for i, sheet in enumerate(xls.sheet_names, 1):
        print(f"  {i}. {sheet}")
else:
    print("Excel file not found, using CSV only")
```

```python
# Load all sheets from Excel file (if it exists)
if grossi_excel.exists():
    sheets_data = {}
    for sheet_name in xls.sheet_names:
        df_sheet = pd.read_excel(grossi_excel, sheet_name=sheet_name)
        sheets_data[sheet_name] = df_sheet
        print(f"\n{sheet_name}: {df_sheet.shape[0]} rows, {df_sheet.shape[1]} columns")
        print(f"  Columns: {df_sheet.columns.tolist()[:5]}...")  # First 5 columns
    
    print(f"\n✓ Loaded {len(sheets_data)} sheets into 'sheets_data' dictionary")
else:
    print("Using CSV data only")
```

## Step 4: Work with Main Data Extraction Sheet

Now that we have the clean data, let's start analyzing it

```python
# Explore the main data columns and values
print("=== MAIN DATA OVERVIEW ===")
print(f"Shape: {df_main.shape}")
print(f"\nColumn names:")
for i, col in enumerate(df_main.columns, 1):
    print(f"  {i}. {col}")

# Check for key columns
print("\n=== KEY COLUMNS ===")
if 'Continent_Ocean' in df_main.columns:
    print(f"Unique Continents/Oceans: {df_main['Continent_Ocean'].nunique()}")
    print(df_main['Continent_Ocean'].value_counts())
else:
    print("Continent_Ocean column not found")
```

```python
# Create visualizations with the main data
from IPython.display import display

if 'Continent_Ocean' in df_main.columns:
    continent_counts = df_main['Continent_Ocean'].value_counts()
    
    fig = px.bar(
        x=continent_counts.index,
        y=continent_counts.values,
        labels={'x': 'Continent/Ocean', 'y': 'Count'},
        title='Distribution by Continent/Ocean - Grossi Data'
    )
    display(fig)

else:    print("Cannot create chart - Continent_Ocean column not found")
```

```python
# Explore threat types
if 'Threat' in df_main.columns:
    threat_counts = df_main['Threat'].value_counts().head(10)
    
    fig = px.bar(
        x=threat_counts.values,
        y=threat_counts.index,
        orientation='h',
        labels={'x': 'Count', 'y': 'Threat Type'},
        title='Top 10 Threat Types'
    )
    display(fig)
else:
    print("Threat column not found")
```

```python
# Explore study design
if 'Study_design' in df_main.columns:
    design_counts = df_main['Study_design'].value_counts()
    
    fig = px.pie(
        values=design_counts.values,
        names=design_counts.index,
        title='Study Design Distribution'
    )
    display(fig)
else:
    print("Study_design column not found")
```

```python
# Explore ecological levels
if 'Ecological_level' in df_main.columns:
    eco_counts = df_main['Ecological_level'].value_counts()
    
    fig = px.bar(
        x=eco_counts.index,
        y=eco_counts.values,
        labels={'x': 'Ecological Level', 'y': 'Count'},
        title='Distribution by Ecological Level'
    )
    display(fig)
else:
    print("Ecological_level column not found")
```
